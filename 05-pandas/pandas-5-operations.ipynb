## Creating another column

Very easy!

````python
## use a fixed value
airbnb['a'] = 100
airbnb.sample(5)
````

````python
## OR we can do a conditionals too

# if overnight rate is over 1000 , we mark it as too expensive

airbnb['too_expensive'] = airbnb['price'] > 1000
airbnb.sample(5)
````

````python
## only find the too expensive ones :-)

too_expensive = airbnb[airbnb['too_expensive'] == True]
print (too_expensive.shape[0])
too_expensive.sample(5)
````

## Sorting Dataframes

We can sort by any columns

````python
## Let's sort by price

airbnb.sort_values(['price'], ascending=True).head(5)
````

````python
# see the most expensive ones first

airbnb.sort_values(['price'], ascending=False).head(5)
````

````python
## We can sort by multiple columns

airbnb.sort_values([ 'number_of_reviews', 'price'], ascending=False).head(5)
````

---

## Groupby Aggregates

let's find min/max/mean prices for each borough

````python
airbnb.groupby('neighbourhood_group').agg({'price': ['mean', 'min', 'max']})
````
