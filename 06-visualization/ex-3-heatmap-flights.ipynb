# Exercise: HeatMap

```python
import pandas as pd
import seaborn as sns
```

```python
flights = sns.load_dataset("flights")
flights
```

```python
# pivoit table

summary = flights.pivot("month", "year", "passengers")

summary
```

```python
# Basic heatmap
sns.heatmap(summary, cmap="BuPu")
```
